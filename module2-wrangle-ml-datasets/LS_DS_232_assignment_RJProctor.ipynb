```python
# prepare input data
# encode output data for train and validation sets
from sklearn.preprocessing import OrdinalEncoder

def prepare_inputs(X_train, X_val):
  # variables have natural order that impact model - use oe to encode
	oe = OrdinalEncoder()
  # fit encoder to training dataset
	oe.fit(X_train)
  # transform training dataset
	X_train_enc = oe.transform(X_train)
  # transform validation dataset
	X_val_enc = oe.transform(X_val)
	return X_train_enc, X_val_enc
```
